```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Titanic-Dataset.csv')
dataset.info()
```

```python
def replace_sex(sex):
    d = {'male':0, 'female': 1}
    return d[sex]

dataset['Sex'] = dataset['Sex'].apply(replace_sex)
```

```python
dataset
```

```python
dataset['Age'].fillna(dataset.Age.mean(), inplace = True)

dataset.Age.isnull().value_counts()
```

```python
dataset.info()
dataset
```

```python
dataset['Embarked'] = dataset.Embarked.replace('S',-1).replace('C',0).replace('Q',1)
```

```python
dataset.plot(kind = 'scatter', x = 'Embarked', y = 'Survived',alpha = 0.005)
```

```python
def cabin_to_num(cabin):

    d = {'A':1, 'B':2, 'C':3,'D':4,'E':5,'G':6,'F':7,'T':8}
    if type(cabin) != float:
        return d[cabin[0]]

dataset['Cabin'] = dataset.Cabin.apply(cabin_to_num)
dataset
```

```python
dataset
```

```python
dataset['Embarked'] = dataset['Embarked'].fillna(dataset['Embarked'].mode()[0])
```

```python
dataset
```

```python
y = dataset['Survived']
x = dataset.drop('Survived', axis = 1)
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=2000)

model.fit(xtrain,ytrain)
```

```python
preds = model.predict(xtest)
from sklearn.metrics import accuracy_score

accuracy_score(ytest, preds)
```

```python
import joblib

joblib.dump(model, 'Titanic_Survival_model.pkl')
```

```python
mymodel = joblib.load('Titanic_Survival_model.pkl')
pred = mymodel.predict(xtest)
accuracy_score(pred, ytest)
```
